```python
import flickrapi
import urllib.request
import os
import time

# Set up the API keys
api_key = ''
api_secret = ''

# Set up the Flickr API
flickr = flickrapi.FlickrAPI(api_key, api_secret, format='parsed-json')

# Set up the search parameters
search_params = {
    'text': 'ai generated art',
    'sort': 'relevance',
    'per_page': 500, # Increase this number to download more photos per page
    'safe_search': 0
}

# Search for photos
result = flickr.photos.search(**search_params)
photos = result['photos']['photo']
pages = result['photos']['pages']
total = result['photos']['total']

print('Found', total, 'photos.')

# Download the photos from all pages
for page in range(1, pages+1):
    print(page)
    search_params['page'] = page
    result = flickr.photos.search(**search_params)
    photos = result['photos']['photo']
    for photo in photos:
        # Get the list of existing files in the images directory
        existing_files = os.listdir('flickr')
        existing_ids = [f.split('_')[0] for f in existing_files]
        if photo['id'] not in existing_ids:
            url = 'https://live.staticflickr.com/{}/{}_{}.jpg'.format(photo['server'], photo['id'], photo['secret'])
            timestamp = int(time.time())  # Get the current time as an integer timestamp
            filename = 'flickr/{}_{}.jpg'.format(photo['id'], timestamp)
            try:
                urllib.request.urlretrieve(url, filename)
            except Exception as e:
                time.sleep(20)
                print('Failed to download', filename, ':', e)
            continue

print('Done.')
```

```python
import os

def delete_duplicate_images(dir_path):
    # create a dictionary to store images by their "before underscore" names
    images_dict = {}
    
    # iterate over all files in the directory
    for file_name in os.listdir(dir_path):
        # get the "before underscore" name
        before_underscore = file_name.split("_")[0]
        
        # check if this name is already in the dictionary
        if before_underscore in images_dict:
            # delete the duplicate file
            os.remove(os.path.join(dir_path, file_name))
        else:
            # add the file to the dictionary
            images_dict[before_underscore] = file_name

# example usage
#delete_duplicate_images("flickr")
```
